### Visualizing Gene Expression Distribution through Heatmap

# Task-2 PCA

```python
from sklearn import decomposition
pca=decomposition.PCA()
from matplotlib.pyplot import figure
```

```python
pca.n_components=183
pca_data=pca.fit_transform(imputed_df)

percentage_var_explained=pca.explained_variance_/np.sum(pca.explained_variance_)

cum_var_explained=np.cumsum(percentage_var_explained)

plt.figure(1,figsize=(6,4))

plt.clf()
plt.plot(cum_var_explained,linewidth=2)
plt.axis('tight')
plt.grid()
plt.xlabel('n_components')
plt.ylabel('Cumulative Explained Variance')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
trans=np.transpose(imputed_df.values)

imputed_df_scaled = StandardScaler().fit_transform(trans)

df_reduced=pca.fit_transform(imputed_df_scaled)

#Finding the index at which adenocarnimo and neuroendocrine tumors occur
hist_para=df_meta['histological_type_other'].values

adenocarnimo_ouccurence=[]
neuroendocrine_occurence=[]

for i in range(imputed_df.shape[1]):
	if str(type(hist_para[i])):
		if ('adenocarcinoma' in str(hist_para[i])):
			adenocarnimo_ouccurence.append(i)
		elif ('neuroendocrine' in str(hist_para[i])):
			neuroendocrine_occurence.append(i)
		
```

```python
PCA_dim1=df_reduced[:,0]
PCA_dim2=df_reduced[:,1]

plt.scatter(PCA_dim1[adenocarnimo_ouccurence],PCA_dim2[adenocarnimo_ouccurence],color='blue')
plt.scatter(PCA_dim2[neuroendocrine_occurence],PCA_dim2[neuroendocrine_occurence],color='orange')
plt.title('PCA Visualization')
plt.grid()
plt.xlabel('PCA_dim1')
plt.ylabel('PCA_dim2')
plt.legend(('Adenocarcinoma Tumor','NeuroEndocrine tumors'))
```

PCA_dim1 has more importance than PCA_dim2

Through this PCA Visualization we can easily separate Adenocarcinoma Tumors from NeuroEndocrine Tumors

```python
print("Variance explained: ",np.sum(pca.explained_variance_ratio_))
```

Thus with these two tumors only 25.29% of our features are explained
